```python
#Write an SQL query to find the position of the alphabet (‘a’) in the first name column ‘Amitabh’ from Worker table.
```

```python
import mysql.connector as connection

try:
    mydb = connection.connect(host="localhost",user="root",passwd="Rudr@123",use_pure=True,database="database_name")
    print(mydb.is_connected())
    cursor = mydb.cursor()
    query = "Select INSTR(First_name, BINARY'a') from worker where First_name = 'Amitabh'"
    cursor.execute(query)
except Exception as e:
    mydb.close()
    print(str(e))
```

```python
#Write an SQL query to print the name of employees having the highest salary in each department.
```

```python
import mysql.connector as connection 

try:
    mydb = connection.connect(host="localhost",user="root",passwd="Rudr@123",use_pure=True,database="database_name")
    print(mydb.is_connected())
    cursor = mydb.cursor()
    query = "Select Dept,Name,Salary from employee where (Dept,Salary) IN (Select Dept,MAX(Salary) from Employee group by Dept)"
    cursor.execute(query)
except Exception as e:
    mydb.close()
    print(str(e))
```
